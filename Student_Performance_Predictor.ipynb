# STEP 1 : Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

# STEP6 :Data Preprocessing

```python
df
```

```python
le = LabelEncoder()
df["Gender"] = le.fit_transform(df["Gender"])
df["Tutoring"] = le.fit_transform(df["Tutoring"])
df["Region"] = le.fit_transform(df["Region"])
df["Parent Education"] = le.fit_transform(df["Parent Education"])
print(le.classes_)
```

```python
df
```

 # STEP 7:Label feature

```python
x = df.drop(["Exam_Score"],axis=1)
y = df["Exam_Score"]
```

# STEP 8:train and testing

```python
x_train,x_test,y_train,y_test = train_test_split(
    x,y,test_size=0.2,random_state=42
)
x_train
x_test
```

# STEP 9: StandardScaler

```python
scaler = StandardScaler()
x_s_train = scaler.fit_transform(x_train)
x_s_test = scaler.transform(x_test)
```

# STEP 10:Model Training

```python
models = {
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest": RandomForestRegressor(),
    "K-Nearest Neighbors": KNeighborsRegressor()
}
```

```python
for name,model in models.items():
  model.fit(x_s_train,y_train)
  y_pred = model.predict(x_s_test)
  r2 = r2_score(y_test, y_pred)
  mae = mean_absolute_error(y_test, y_pred)
  mse = mean_squared_error(y_test, y_pred)
  rmse = np.sqrt(mse)
```

```python
best_model = None
best_r2 = -1
if r2 > best_r2:
    best_r2 = r2
    best_model = model

print(f"Best Model: {best_model}")
print(f"Best R2 Score: {best_r2:.4f}")
```

# STEP 11:RESULT

```python
print(f"\n{name}")
print(f"R² Score : {r2:.4f}")
print(f"MAE      : {mae:.4f}")
print(f"MSE      : {mse:.4f}")
print(f"RMSE     : {rmse:.4f}")
```

# STEP 12:GridSearchCV

```python
param_grid = {
    'n_neighbors':[ 3,4,5,7,11],
    'weights': ["uniform", "distance"],
    'metric': ["euclidean", "manhattan"]
}
```

```python
grid = GridSearchCV(
    KNeighborsRegressor(),
    param_grid,
    cv=5,
    scoring="r2"
)
grid.fit(x_s_train,y_train)
print("Best Parameters:", grid.best_params_)
print("Best Score:", grid.best_score_)
```
